```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np

sys.path.append(str(Path.cwd().parent / "scripts"))

import config
import cleaning as clean
```

```python
df = pd.read_csv(config.DATA_RAW / "financials.csv")
df["Company"].unique()
# df["Section"].unique()
# df
```

```python
companies = [
    "HUL",
    "ITC",
    "Dabur",
    "Marico",
    "Colgate",
    "Nestle",
    "Godrej",
    "Emami"
]
companies = clean.split_by_column(df, "Company")
companies["Nestle"].head(10) 
```

```python
# drop the now-redundant grouping column from each company sub_df (logic in cleaning.py)
for company, sub_df in companies.items():
    clean.drop_column(sub_df, "Company")
    clean.drop_column(sub_df, "ValueRaw")

companies["Nestle"].head(10)
```

```python
companies = {name: clean.split_by_column(sub_df, "Section") for name, sub_df in companies.items()}
companies["Nestle"]["P&L"].head(10)
```

```python
# drop the redundant Section column from every innermost sub_df (logic in cleaning.py)
for company, sections in companies.items():
    for section, sub_df in sections.items():
        clean.drop_column(sub_df, "Section")

companies["Nestle"]["P&L"].head(10)
```

```python
# add key = "<LineItem>_<Period>" (e.g. 'Sales_FY2024') to every innermost sub_df
for company, sections in companies.items():
    for section, sub_df in sections.items():
        clean.add_key_column(sub_df)

companies["Nestle"]["P&L"].head(10)
```

```python
# reshape each innermost sub_df to wide: LineItem -> columns, Period(12) -> rows, Value -> cells
# (no period filter -> keep all 12 periods FY2015..FY2026; logic in cleaning.py)
wide = {company: {section: clean.pivot_line_items(sub_df)
                  for section, sub_df in sections.items()}
        for company, sections in companies.items()}

wide["Nestle"]["BalanceSheet"]
```

```python
# save each wide frame -> data/cleaned/<Company>/<Section>.csv  (8 folders x 3 files = 24; logic in cleaning.py)
clean.save_wide(wide, config.DATA_CLEANED)
```
